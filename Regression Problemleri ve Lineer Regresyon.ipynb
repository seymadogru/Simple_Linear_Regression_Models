```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
house_prices_df = pd.read_csv("house_prices_train.csv")
house_prices_df.head()
```

Önemli olabilecek feature'lar;

-YearBuilt: Original construction date (orijinal yapım tarihi)

-LotArea: Lot size in square feet (arsa büyüklüğü)

-Heating: Type of heating (ısıtma türü)

-Bedroom: Number of bedrooms above basement level (yatak odası sayısı)

-TotRmsAbvGrd: Total rooms above grade (does not include bathrooms) (toplam oda sayısı)

-FullBath: Full bathrooms above grade (tam banyolar)

-HalfBath: Half baths above grade (yarım banyolar)

-GarageCars: Size of garage in car capacity (garajın araç kapasitesi)

-SalePrice - the property's sale price in dollars. This is the target variable that you're trying to predict.

```python
new_features = house_prices_df[["YearBuilt","LotArea","TotRmsAbvGrd","FullBath","HalfBath","GarageCars","SalePrice"]]
```

```python
new_features
```

```python
Y = new_features['SalePrice']
X = new_features[['YearBuilt','LotArea','TotRmsAbvGrd','FullBath','HalfBath','GarageCars']]
```

```python
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)
```

```python
print('Değişkenler: \n', lrm.coef_)
print('Sabit değer (bias): \n', lrm.intercept_)
```

```python
import statsmodels.api as sm

X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```

R-squared değeri yüzde 60 çıktı. Yani ortalama bir tahmin yaptık bu feature'larla. Seçtiğimiz feature'lardan halfbath hariç hepsi fiyatı etkileyen unsurlarmış. 

Adj. R-squared değerinin de düşük çıkmasının sebebi, feature'larımızda korele olan değerler var demek.
